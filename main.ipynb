```python
from ingestion import ingest_logs
from pattern_recognition import extract_patterns
from anomaly_detection import detect_anomalies
from incident_prediction import predict_incident
from alerting import raise_alert

def main():
    logs = ingest_logs("logs.csv")
    patterns = extract_patterns(logs)
    anomalies = detect_anomalies(patterns)

    if predict_incident(anomalies):
        raise_alert(logs, anomalies)
    else:
        print("âœ… System stable. No incidents predicted.")

if __name__ == "__main__":
    main()
```
